## Ćwiczenie 3
### Znajdź najstarszy i najmłodszy film do wypożyczenia.

```python
ex3 = pd.read_sql("""(SELECT 
                        film_id,
                        title,
                        release_year
                    FROM 
                        film 
                    ORDER BY 
                        release_year ASC LIMIT 1)

                    UNION ALL 

                    (SELECT 
                        film_id,
                        title,
                        release_year
                    FROM 
                        film 
                    ORDER BY 
                        release_year DESC LIMIT 1)""", con=connection)
ex3
```

## Ćwiczenie 4
### Ile wypożyczeń odbyło się między 2005-07-01 a 2005-08-01?

```python
ex4 = pd.read_sql("""SELECT 
                        COUNT(*) 
                    FROM 
                        rental 
                    WHERE 
                        rental_date 
                    BETWEEN 
                        '2005-07-01' AND '2005-08-01' """,con=connection)
ex4
```

## Ćwiczenie 7

```python
ex7 = pd.read_sql("""SELECT 
                        customer_id, 
                        first_name, 
                        last_name 
                    FROM 
                        customer 
                    JOIN 
                        address USING(address_id) 
                    JOIN 
                        city USING(city_id) 
                    JOIN 
                        country USING(country_id) 
                    WHERE 
                        country.country IN ('Poland', 'Nigeria', 'Bangladesh')""", con= connection)
ex7
```

## Ćwiczenie 11
### Znajdź wszystkie kategorie filmów wypożyczonych w Ameryce

```python
ex11 = pd.read_sql("""SELECT
                        DISTINCT
                        name, 
                        country 
                    FROM 
                        category 
                    JOIN 
                        film_category USING(category_id) 
                    JOIN 
                        film USING(film_id) 
                    JOIN 
                        inventory USING(film_id) 
                    JOIN 
                        store USING(store_id)
                    JOIN 
                        address USING(address_id) 
                    JOIN 
                        city USING(city_id) 
                    JOIN 
                        country USING(country_id) 
                    WHERE 
                        country = 'United States' """, con= connection)
ex11
```

```python
ex11_c = pd.read_sql("""SELECT
                        COUNT(*)
                    FROM 
                        category 
                    JOIN 
                        film_category USING(category_id) 
                    JOIN 
                        film USING(film_id) 
                    JOIN 
                        inventory USING(film_id) 
                    JOIN 
                        store USING(store_id)
                    JOIN 
                        address USING(address_id) 
                    JOIN 
                        city USING(city_id) 
                    JOIN 
                        country USING(country_id) 
                    WHERE 
                        country = 'United States' """, con= connection)

ans = ex11_c['count'][0]
print(f"W Ameryce wypożyczono {ans} kategorii filmów.")
```
